RENAMING OUR COLUMNS FOR UNIFORMITY

```
#renaming the columns PRODUCT and DATETIME in Telcom_df
Telcom_df.rename(columns = {'PRODUTC':'PRODUCT', 'DATETIME':'DATE_TIME'}, inplace = True)
Telcom_df.head()
```

```
#renaming the columns DW_A_NUMBER and DW_B_NUMBER in Telcom_df2
Telcom_df2.rename(columns = {'DW_A_NUMBER':'DW_A_NUMBER_INT', 'DW_B_NUMBER':'DW_B_NUMBER_INT'}, inplace =True)
Telcom_df2.head()
```

```
#renaming the columns CELLID and SEIT_ID in Telcom_df3
Telcom_df3.rename(columns = {'CELLID':'CELL_ID', 'SIET_ID':'SITE_ID'}, inplace = True)
Telcom_df3.head()
```

CONCATATING OUR DATASETS

```
#merging our datasets from the three days into one table for ease of manipulation
Telcom_all_df = pd.concat((Telcom_df, Telcom_df2, Telcom_df3), axis = 0, ignore_index = True)
print(Telcom_all_df)

```

DROPPING ROWS AND COLUMNS NOT REQUIRED IN OUR EXAMPLE

```
#our dataset contains columns that are not useful for this projecs
#we need to drop them
Telcom_all_df.drop(['CELL_ON_SITE','COUNTRY_A', 'COUNTRY_B'], axis = 1, inplace = True)
Telcom_all_df

```

MERGING OUR TABLES TO ONE DATAFRAME

```
# combining the cells geo dataset to the new Telcom dataset
New_df =Telcom_all_df.merge(cells_geo_columns , how = 'left', on = 'CELL_ID')
New_df
New_df.drop_duplicates(['CELL_ID'])
```
